```python
A = df["churn"].value_counts()
print(A)
```

```python
df["churn"].dtype
```

```python
plt.pie(A, labels=["Not churned", "churned"], colors=["orange", "lime"], startangle=50, shadow=True, radius=2, explode=(0,0.2), autopct="%1.2f%%", pctdistance=0.75)
circle= plt.Circle((0,0), 1,color="white")
c=plt.gcf()
c.gca().add_artist(circle)
plt.title("Churned vs Not-Churned customers")
 

plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.countplot(x="state", data=df)
plt.xticks(fontsize=14)
plt.yticks(fontsize=13)

plt.xlabel("States", fontsize=20, color="Red")
plt.ylabel("Count", fontsize=20, color="Green")

plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(20,7))

sns.countplot(x="state", data=df, hue=df.churn)

plt.xlabel("States", fontsize=20, color="Red")
plt.ylabel("Count", fontsize=20, color="Green")

plt.show()
```
